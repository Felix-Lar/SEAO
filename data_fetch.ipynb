```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET


tree = ET.parse('Data/2015/avis.xml')
root = tree.getroot()



# store data in these dictionaries
avis = {}
avis_soum = {}

#fetch
avis_tot = len(root)
avis_num = 0
for i in range(avis_tot):
    att_num = len(root[i])
    job = {}
    
    for j in range(att_num):
        if root[i][j].tag != 'fournisseurs':
            job[root[i][j].tag] = root[i][j].text
        if root[i][j].tag == 'fournisseurs':
            fourn_tot = len(root[i][j])
            fourn_num = 0
            fourn = {}
            for k in range(fourn_tot):
                info_fourn = {}
                fourn_att_num = len(root[i][j][k]) 
                for l in range(fourn_att_num):
                    info_fourn[root[i][j][k][l].tag] = root[i][j][k][l].text


                fourn[fourn_num] = info_fourn
                fourn_num +=1

    avis[avis_num] = job
    avis_soum[avis_num] = fourn
    avis_num +=1



#jobs
df1 = pd.DataFrame(avis).T
display(df1)


#soumissionaires pour job #0

job_pos = 20

df2 = pd.DataFrame(avis_soum[job_pos]).T
display(df2)

```
